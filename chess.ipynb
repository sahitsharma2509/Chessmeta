## To Determine the top 20 Highest Winrates of Players with games played >=50 

```python
pdict = {}
pwins = []
pgames = []
for player in chess_data['White'].unique():
    
    pwins.append(sum(chess_data[chess_data["White"]==player]['Result'] =='White')+sum(chess_data[chess_data["Black"]==player]['Result']=='Black'))
    pgames.append(len(chess_data[chess_data["White"]==player]['Result'] =='White')+len(chess_data[chess_data["Black"]==player]['Result']=='Black'))
    
        
        
```

```python
pwrate = pd.DataFrame(list(zip(chess_data['White'].unique(),pwins,pgames)),columns=["Username","No. of Wins","Total Games Played"])
```

```python
pwrate["Winrate"] = pwrate["No. of Wins"]*100/pwrate["Total Games Played"]
```

```python
pwrate = pwrate.sort_values("Winrate",ascending = False)
```

```python
pwrate_20 = pwrate[pwrate["Total Games Played"]>50][:20]
```

```python
plt.figure(figsize=(30,15))
sns.set_context('poster',font_scale=1.5)
plt.title("Highest winrate, Games played >=50")
sns.barplot(y="Username",x="Winrate",data=pwrate_20)
```
